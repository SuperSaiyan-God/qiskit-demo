```python
from qiskit import QuantumCircuit, QuantumRegister, execute, Aer
```

```python
qc = QuantumCircuit(14)
```

```python
i = 0
while(i<12):
    qc.cx(i,i+2)
    qc.cx(i+1,i+2)
    qc.x(i+2)
    qc.x(i+1)
    qc.ccx(i,i+1,i+3)
    qc.x(i+1)
    i+=4
```

```python
qc.draw(output='mpl')
```
