```python
q0 = QuantumRegister(5, 'q0')

switch = QuantumCircuit(q0, name="Switch")
```

```python
switch.reset(q0[4])
switch.cx(q0[0],q0[1])
switch.cx(q0[2],q0[0])
switch.cx(q0[0],q0[1])
switch.barrier(q0[1])
switch.cx(q0[0],q0[1])
switch.ccx(q0[2],q0[1],q0[4])
switch.cx(q0[3],q0[0])
switch.cx(q0[0],q0[1])
switch.cx(q0[4],q0[0])

switch.draw(output='mpl')
```
